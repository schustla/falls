```python
#中位数IQR
def DM(x):
    result = np.around(np.median(x),decimals=2).astype(str)+'('+ np.around(np.percentile(x, 25),decimals=2).astype(str)+', '+np.around(np.percentile(x, 75),decimals=2).astype(str)+')'
    return result
    
df2=pd.pivot_table(data=df1
                   ,index=[u'LABEL'],values=[u'BMXBMI']
                   ,aggfunc=DM,
              )
df2
```

```python
#均值标准差
def DM(x):
    result = np.around(np.mean(x),decimals=2).astype(str)+'('+np.around(np.std(x),decimals=2).astype(str)+')'
    return result

df2=pd.pivot_table(data=df1
                   ,index=[u'year'],values=[u'URXUBA',u'URXUCD',u'URXUCO',u'URXUCS',u'URXUMO',u'URXUMN',u'URXUPB',u'URXUSB',u'URXUSN',u'URXUTL',u'URXUTU',u'LBXBPB',u'LBXBCD',u'LBXTHG',u'LBXBSE',u'LBXBMN']
                   ,aggfunc=DM,
              )
df2.T
```

```python
#特征
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
d1 = df1['RIAGENDR'].values.reshape(-1, 1)
d2 = df1['RIDRETH3'].values.reshape(-1, 1)
d3 = df1['DMDEDUC2'].values.reshape(-1, 1)
df1['RIAGENDR'] = OneHotEncoder(categories='auto').fit_transform(d1).toarray()
df1['RIDRETH3'] = OneHotEncoder(categories='auto').fit_transform(d2).toarray()
df1['DMDEDUC2'] = OneHotEncoder(categories='auto').fit_transform(d3).toarray()
```

```python
df1=df1.dropna()
```

```python
X = df1.iloc[:,1:23]
y = df1['LABEL']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
data = X
scaler = StandardScaler() #实例化
scaler.fit(data) #fit，本质是生成均值和方差
scaler.mean_ #查看均值的属性mean_
scaler.var_ #查看方差的属性var_
x_std = scaler.transform(data) #通过接口导出结果
x_std.mean() #导出的结果是一个数组，用mean()查看均值
x_std.std() #用std()查看方差
scaler.fit_transform(data) #使用fit_transform(data)一步达成结果
scaler.inverse_transform(x_std) #使用inverse_transform逆转标准化

X = scaler.fit_transform(data)
```

```python
y
```

```python
#PCA
pca = PCA(n_components=19)
pca = pca.fit(X) 
X_dr = pca.transform(X) 
```

```python
#属性explained_variance，查看降维后每个新特征向量上所带的信息量大小（可解释性方差的大小）
pca.explained_variance_
```

```python
#属性explained_variance_ratio，查看降维后每个新特征向量所占的信息量占原始数据总信息量的百分比
#又叫做可解释方差贡献率
pca.explained_variance_ratio_
```

```python
#大部分信息都被有效地集中在了第一个特征上
pca.explained_variance_ratio_.sum()
```

```python
pca_f = PCA(n_components=0.95,svd_solver="full")
pca_f = pca_f.fit(X)
X_f = pca_f.transform(X)
pca_f.explained_variance_ratio_
```

```python
#SKB
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif ## 导入 f_classif 检验
from sklearn.feature_selection import SelectKBest,chi2
skb = SelectKBest(f_classif, k=18)
X_new=skb.fit_transform(X, y)
pd.DataFrame(X_new).head(10)
```
